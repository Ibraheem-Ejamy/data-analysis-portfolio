The dataset includes movies and TV shows available on Netflix, with attributes such as title, director, cast, country, release year, duration, and genre.  
For this project, we'll mainly focus on movies from the 1990s.


```python
# Importing pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df
```

## üé¨ Finding Short Action Movies

Let's find out how many action movies released during the 1990s were considered 'short' (less than 90 minutes).


```python
# Start coding here! Use as many cells as you like
duration = int((netflix_df.loc[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999), 'duration']).value_counts().idxmax())

short_movie_count = 0
gen_re = netflix_df[netflix_df['genre'] == "Action"]
for lab, row in gen_re.iterrows():
    if row['duration'] < 90 and row['release_year'] in range(1990, 1999):
        short_movie_count += 1

print(f"Number of short Action movies (1990s): {short_movie_count}")
```

### ‚ú® Insight
We found **7 short action movies** from the 1990s, showing that shorter action films were not very common during this decade.


## üìÖ Movie Releases per Year

Let's visualize the number of movies released each year during the 1990s.


```python
plt.figure(figsize=(24,10))
sns.countplot(data=netflix_df[netflix_df['release_year'].between(1990, 1999)], 
              x='release_year', palette='viridis')
plt.title("Number of Netflix Movies Released per Year (1990s)")
plt.xlabel("Year")
plt.ylabel("Movie Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ‚ú® Insight
Some years had a spike in movie releases. This might reflect Netflix‚Äôs acquisition strategy or production trends.


## ‚è∞ Distribution of Movie Durations

Now, let's check the distribution of movie lengths during the 90s.


```python
plt.figure(figsize=(26,12))
sns.histplot(netflix_df[netflix_df['release_year'].between(1990, 1999)]['duration'], 
             bins=20, kde=True, color = 'seagreen')
plt.title("Distribution of Movie Durations (1990s)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

### ‚ú® Insight
Most movies tend to have durations around **[100]** minutes, suggesting standard viewer preferences.


## üé≠ Top 10 Genres in 1990s Netflix Movies

Which genres ruled the 90s? Let‚Äôs find out!


```python
genre_90s = netflix_df[netflix_df['release_year'].between(1990, 1999)]['genre'].value_counts().head(10)
genre_90s.plot(kind='bar', color='teal', figsize=(24,10))
plt.title("Top 10 Genres (1990s Netflix Movies)")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ‚ú® Insight
Genres like **[Action - Dramas - Comedies]** dominated the 90s Netflix catalog, reflecting what audiences loved back then.
