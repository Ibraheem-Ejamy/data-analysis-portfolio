## üßæ About the Dataset

We‚Äôre using a dataset inspired by a paper titled **"Climate Change Adaptation Innovation in the Water Sector in Africa"**. It contains information from the years 1990, 2000, 2005, and 2010 to 2016.

Here's what's inside:
- `year`
- `country`
- `water_related_adaptation_tech`
- `trade` (trade openness % of GDP)
- `days_to_register_property`
- `gdp_per_capita`

- `water_stress_index`

## üîç SQL Exploration Steps

### 1. View the Full Dataset

```python
-- Explore the data in the table
SELECT *
FROM climate
```

### 2. Countries with Highest Water Stress

```python
SELECT country, water_stress_index
FROM climate
ORDER BY 2 DESC
```

### 3. Countries with Lowest GDP per Capita

```python
SELECT DISTINCT country, year, gdp_per_capita
FROM climate
ORDER BY gdp_per_capita
LIMIT 10;
```

### 4. Countries with Moderate Water Stress

```python
SELECT country, year, water_stress_index
FROM climate
WHERE water_stress_index BETWEEN 0.5 AND 0.6;
```

### 5. Filter by Country Name and Stress Level

```python
SELECT country, water_stress_index
FROM climate
WHERE water_stress_index > 0.5
AND (country LIKE 'E%'  OR country LIKE 'S%');
```

## Aggregate, group, and sort the data

### 6. Average Adaptation Technology by Country

```python
SELECT country, AVG(water_related_adaptation_tech) AS avg_water_related_adaptation_tech
FROM Climate
GROUP BY country
ORDER BY avg_water_related_adaptation_tech DESC;
```

### 7. Countries with High Adaptation Tech Use

```python
SELECT country
FROM climate
GROUP BY country
HAVING AVG(water_related_adaptation_tech) > 1;
```

## üìä Data Visualizations in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./climate.csv")
```

### 1. Average Adaptation Tech by Country

```python
# Compute average adaptation tech
avg_adapt = data.groupby('country')['water_related_adaptation_tech'].mean().sort_values(ascending=False)

plt.figure(figsize=(24, 10))
bars = avg_adapt.plot(kind='bar', color='teal')
plt.title('Average Water-Related Adaptation Technologies by Country')
plt.ylabel('Avg. Adaptation Tech')
plt.xticks(rotation=45, ha='right')

for i, value in enumerate(avg_adapt):
    bars.text(i, value + 0.05, f'{value:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

### 2. Water Stress Index Over Time

```python
top_countries = data.groupby('country')['water_stress_index'].mean().sort_values(ascending=False).head(5).index
subset = data[data['country'].isin(top_countries)]

plt.figure(figsize=(20, 9))
sns.lineplot(data=subset, x='year', y='water_stress_index', hue='country', marker='o')
plt.title('Water Stress Index Over Time (Top 5 High-Stress Countries)')
plt.ylabel('Water Stress Index')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
# Define regions manually
region_map = {
    'Algeria': 'North Africa', 'Angola': 'Central Africa', 'Benin': 'West Africa', 'Botswana': 'Southern Africa',
    'Burkina Faso': 'West Africa', 'Burundi': 'East Africa', 'Cabo Verde': 'West Africa', 'Cameroon': 'Central Africa',
    'CAR': 'Central Africa', 'Chad': 'Central Africa', 'Comoros': 'East Africa', "Cote d'Ivoire": 'West Africa',
    'Djibouti': 'East Africa', 'Egypt': 'North Africa', 'Equatorial Guinea': 'Central Africa', 'Eritrea': 'East Africa',
    'Eswatini': 'Southern Africa', 'Ethiopia': 'East Africa', 'Gabon': 'Central Africa', 'Gambia': 'West Africa',
    'Ghana': 'West Africa', 'Guinea': 'West Africa', 'Kenya': 'East Africa', 'Lesotho': 'Southern Africa',
    'Liberia': 'West Africa', 'Libya': 'North Africa', 'Madagascar': 'East Africa', 'Malawi': 'East Africa',
    'Mali': 'West Africa', 'Mauritania': 'West Africa', 'Mauritius': 'East Africa', 'Morocco': 'North Africa',
    'Mozambique': 'East Africa', 'Namibia': 'Southern Africa', 'Niger': 'West Africa', 'Nigeria': 'West Africa',
    'Rwanda': 'East Africa', 'Sao Tome and Principe': 'Central Africa', 'Senegal': 'West Africa', 'Seychelles': 'East Africa',
    'Sierra Leone': 'West Africa', 'Somalia': 'East Africa', 'South Africa': 'Southern Africa', 'Sudan': 'North Africa',
    'Tanzania': 'East Africa', 'Togo': 'West Africa', 'Tunisia': 'North Africa', 'Uganda': 'East Africa',
    'Zambia': 'East Africa', 'Zimbabwe': 'Southern Africa'
}

data['region'] = data['country'].map(region_map)
df_region = data[data['region'].notnull()]  # Remove countries not in map

# Create the FacetGrid WITHOUT legend
g = sns.FacetGrid(df_region, col="region", col_wrap=2, height=6, aspect=2, sharey=False)

# Plot each lineplot with hue
g.map_dataframe(sns.lineplot, x="year", y="water_related_adaptation_tech", hue="country", marker="o")

# Turn off automatic legend generation
for ax in g.axes.flat:
    ax.legend_.remove() if ax.get_legend() else None

# For each axis (subplot), add its own mini-legend
for ax, (region_name, region_data) in zip(g.axes.flat, df_region.groupby('region')):
    handles, labels = ax.get_legend_handles_labels()
    if handles:  # Only add legend if there are plots
        ax.legend(handles, labels, title="Country", loc='upper center', bbox_to_anchor=(0.5, -0.25),
                  ncol=6, fontsize='medium', title_fontsize='medium', frameon=False)

# Titles, labels
g.set_titles("{col_name}")
g.set_axis_labels("Year", "Adaptation Tech")

# Set figure size
g.fig.set_size_inches(28, 20)

plt.tight_layout()
plt.show()
```

## üìå Key Insights from the Data
After performing exploratory data analysis on the climate adaptation dataset, here are the most important findings:

### üåü 1. South Africa, Egypt, and Morocco Lead in Adaptation Tech
- These three countries have the highest average usage of water-related adaptation technologies.

- This reflects stronger institutional capacity or investment in climate adaptation.

### üåä 2. High Water Stress Isn‚Äôt Always Tied to Wealth
- Countries like Kenya, Eswatini, and Nigeria experience severe water stress.

- However, they don‚Äôt always have a high GDP per capita, which shows that climate vulnerability can exist regardless of economic strength.

### ‚ö†Ô∏è 3. Moderate Stress + Low Tech = Risk Zones
- Countries such as Somalia and Kenya consistently show moderate to high water stress, yet display low adaptation tech levels.

- These areas might benefit from targeted support or international aid.

### üåç 4. Regional Trends Are Clear
- North African and Southern African regions have higher tech adoption.

- West, East, and Central African countries generally have lower averages, suggesting potential regional inequality in innovation capacity.

### üö® 5. Only a Few Countries Have High Tech Usage
- Just three countries‚ÄîSouth Africa, Egypt, and Morocco‚Äîhave an average adaptation tech score greater than 1.

- Most other countries fall below this, indicating that tech diffusion is still limited across the continent.
